```python
#importing libraries used

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
#Function used to supress the warnings 

def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
warnings.filterwarnings('ignore')
```

```python
#Analyse the data (from the dataset )

df = pd.read_csv('auto-mpg1.csv')
df.sample(5)
df.shape
df.plot.scatter(x ='horsepower', y = 'mpg')
df.plot.bar(x ='horsepower', y = 'weight')
```

```python
# Setting target variable for model prediction (mileage)
# Set the values from which model learns (horespower and weight)

target = df['mpg']
features = df[['horsepower','weight']]
```

```python
# Train the linear regression model 

lr = LinearRegression()
lr.fit(features,target)
```

```python
# Obtain the model score

lr.score(features,target)
```

```python
#Testing this model to predict a value based on horsepower and weight 

hp=100
weight=2000
pred=lr.predict([[hp,weight]])

print("The predicted mileage for horsepower "+str(hp)+" and weight "+str(weight)+" is: "+str(pred[0]))
```
